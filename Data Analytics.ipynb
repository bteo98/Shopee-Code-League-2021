```python
import json
```

```python
with open("./contacts.json") as filepath:
    data = json.loads(filepath.read())
```

```python
def update(old, new, dic, reversedic):
    for Id in reversedic[old]:
        dic[Id] = new
    reversedic[new].extend(reversedic[old])
```

```python
email = {}
reverseEmail = {}
phone = {}
reversePhone = {}
orderId = {}
reverseOrderId = {}
namelist = {}
names = []

for ticket in data:    
    if ticket['Email'] != '' and ticket['Email'] in email:
        name = email[ticket['Email']]
        namelist[name][1] += ticket['Contacts']
        namelist[name][0] = str(namelist[name][0]) + "-" + str(ticket['Id'])
        
        if ticket['Phone'] != '' and ticket['Phone'] not in phone:
            phone[ticket['Phone']] = name
            if name not in reversePhone:
                reversePhone[name] = [ticket['Phone'],]
            else:
                reversePhone[name].append(ticket['Phone'])
                
        elif ticket['Phone'] != '' and ticket['Phone'] in phone:
            other = phone[ticket['Phone']]
            namelist[name][1] += namelist[other][1]
            namelist[name][0] = str(namelist[name][0]) + "-" + str(namelist[other][0])
            update(other, name, email, reverseEmail)
            update(other, name, phone, reversePhone)
            update(other, name, orderId, reverseOrderId)
            del namelist[other]
            
        if ticket['OrderId'] != '' and ticket['OrderId'] not in orderId:
            orderId[ticket['OrderId']] = name
            if name not in reverseOrderId:
                reverseOrderId[name] = [ticket['OrderId'],]
            else:
                reverseOrderId[name].append(ticket['OrderId'])
                
        elif ticket['OrderId'] != '' and ticket['OrderId'] in orderId:
            other = orderId[ticket['OrderId']]
            namelist[name][1] += namelist[other][1]
            namelist[name][0] = str(namelist[name][0]) + "-" + str(namelist[other][0])
            update(other, name, email, reverseEmail)
            update(other, name, phone, reversePhone)
            update(other, name, orderId, reverseOrderId)
            del namelist[other]
            
    elif ticket['Phone'] != '' and ticket['Phone'] in phone:
        name = phone[ticket['Phone']]
        namelist[name][1] += ticket['Contacts']
        namelist[name][0] = str(namelist[name][0]) + "-" + str(ticket['Id'])
        
        if ticket['Email'] != '' and ticket['Email'] not in email:
            email[ticket['Email']] = name
            if name not in reverseEmail:
                reverseEmail[name] = [ticket['Email'],]
            else:
                reverseEmail[name].append(ticket['Email'])
                
        elif ticket['Email'] != '' and ticket['Email'] in email:
            other = email[ticket['Email']]
            namelist[name][1] += namelist[other][1]
            namelist[name][0] = str(namelist[name][0]) + "-" + str(namelist[other][0])
            update(other, name, email, reverseEmail)
            update(other, name, phone, reversePhone)
            update(other, name, orderId, reverseOrderId)
            del namelist[other]
            
        if ticket['OrderId'] != '' and ticket['OrderId'] not in orderId:
            orderId[ticket['OrderId']] = name
            if name not in reverseOrderId:
                reverseOrderId[name] = [ticket['OrderId'],]
            else:
                reverseOrderId[name].append(ticket['OrderId'])
                
        elif ticket['OrderId'] != '' and ticket['OrderId'] in orderId:
            other = orderId[ticket['OrderId']]
            namelist[name][1] += namelist[other][1]
            namelist[name][0] = str(namelist[name][0]) + "-" + str(namelist[other][0])
            update(other, name, email, reverseEmail)
            update(other, name, phone, reversePhone)
            update(other, name, orderId, reverseOrderId)
            del namelist[other]
            
    elif ticket['OrderId'] != '' and ticket['OrderId'] in orderId:
        name = orderId[ticket['OrderId']]
        namelist[name][1] += ticket['Contacts']
        namelist[name][0] = str(namelist[name][0]) + "-" + str(ticket['Id'])
        
        if ticket['Phone'] != '' and ticket['Phone'] not in phone:
            phone[ticket['Phone']] = name
            if name not in reversePhone:
                reversePhone[name] = [ticket['Phone'],]
            else:
                reversePhone[name].append(ticket['Phone'])
                
        elif ticket['Phone'] != '' and ticket['Phone'] in phone:
            other = phone[ticket['Phone']]
            namelist[name][1] += namelist[other][1]
            namelist[name][0] = str(namelist[name][0]) + "-" + str(namelist[other][0])
            update(other, name, email, reverseEmail)
            update(other, name, phone, reversePhone)
            update(other, name, orderId, reverseOrderId)
            del namelist[other]
            
        if ticket['Email'] != '' and ticket['Email'] not in email:
            email[ticket['Email']] = name
            if name not in reverseEmail:
                reverseEmail[name] = [ticket['Email'],]
            else:
                reverseEmail[name].append(ticket['Email'])
                
        elif ticket['Email'] != '' and ticket['Email'] in email:
            other = email[ticket['Email']]
            namelist[name][1] += namelist[other][1]
            namelist[name][0] = str(namelist[name][0]) + "-" + str(namelist[other][0])
            update(other, name, email, reverseEmail)
            update(other, name, phone, reversePhone)
            update(other, name, orderId, reverseOrderId)
            del namelist[other]
        
    else:
        if len(names) == 0:
            names.append(1)
        else:
            names.append(names[-1] + 1)
        
        namelist[names[-1]] = [str(ticket['Id']), ticket['Contacts']]
        email[ticket['Email']] = names[-1]
        reverseEmail[names[-1]] = [ticket['Email'], ]
        phone[ticket['Phone']] = names[-1]
        reversePhone[names[-1]] = [ticket['Phone'], ]
        orderId[ticket['OrderId']] = names[-1]
        reverseOrderId[names[-1]] = [ticket['OrderId'], ]
        
```

```python
import csv

data = namelist
results = {}

for key, value in data.items():
    IDs = value[0]
    contact = value[1]
    IDList = IDs.split("-")
        
    IDList = list(map(lambda x: int(x), IDList))

    IDList = sorted(IDList)
    
    traceString = str(IDList[0])
    for ID in IDList[1:]: 
        traceString = traceString + "-" + str(ID)

    for ID in IDList:
        if ID not in results:
            results[ID] = traceString + ", " + str(contact)
```
